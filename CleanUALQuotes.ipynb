The zeroeth entry is the day's opening quote.  Column zero is the 'time', where the first entry of a given day is given in Unix Epoch time, and the following times count the number of minutes that have passed since the opening of the NYSE that day.

```python
data.columns = ["DATE", "CLOSE", "HIGH", "LOW", "OPEN", "VOLUME"] 
#this is the column order given by Google Finance
```

That's the column order given by Google Finance.  Next, because the time stamps are useless to me in their current form, I'm going to switch them over to Epoch times, which are more widely used.

```python
data["EpochTime"] = np.zeros(len(data.index)) 
```

I need to get rid of the 'a' before the Epoch times and create Epoch times associated to each of the other time stamps. The 'a's only occur at the opening of each day, and actually, they characterize the opening entry.  So I'm using them to build the Epoch time stamps of each entry of their respective day.  Afterwards I place the new time stamp into the "EpochTime" column of the dataframe.

```python
for i in range(len(data.index)) :
    if data["DATE"][i][0]=='a':
        data.loc[i,"EpochTime"]=str(data.loc[i,"DATE"][1:11]) 
        j=i
        #this removes the 'a' from the epoch time stamps
    else :
        data.loc[i,"EpochTime"]=str(int(data["DATE"][i])*60 + int(data["EpochTime"][j]))
        #adds the number of seconds that have passed since the start of the trading
        #to the epoch time stamp at the start of the trading.  This is now the 
        #epoch time stamp for the entry at index i
```
