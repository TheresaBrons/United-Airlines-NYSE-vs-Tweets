```python
import numpy as np
import pandas as pd
import pyflux as pf
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib
import pylab as pl
import scipy.stats as PySTATS
from scipy.stats import norm
```

```python
data = pd.read_csv('/home/theresa/Desktop/DataSets/UnitedTweets/UnitedTweet_sentiment_polarity.csv')

```

```python
data.index=data['created_at']
```

```python
dateTimes=pd.to_datetime(data.index)
```

```python
dateTimes = dateTimes.tz_localize('UTC').tz_convert('US/Eastern') #Twitter time-stamps are in UTC
data['created_at']=dateTimes

data.index=dateTimes
#setting the indices of the dataframe to be datetimes so I can .resample
```

```python
sentiment1m = data['Sentiment'].resample('360t', how='mean')
#I chose to graph the 6 hour average, as the sentiment swings so wildly


```

```python
fig, ax = plt.subplots()
sentiment1m.plot(legend=False,figsize=(10,7), color = 'red', linewidth=.5)

plt.setp(ax.get_xticklabels(),fontsize=14,family='sans-serif')
plt.setp(ax.get_yticklabels(),fontsize=14,family='sans-serif')
plt.xlabel('Time',fontsize=20,family='sans-serif')
plt.ylabel('Tweet Sentiment',fontsize=20,family='sans-serif')

plt.grid(False)
plt.savefig("UnitedTweetSentimentOverTime.png")
plt.show()
```
